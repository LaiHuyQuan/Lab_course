```python
import pandas as pd

path = "./archive/xclara.csv"
df_raw = pd.read_csv(path)

df_raw.shape
```

```python
df = df_raw.copy()
df.head()
```

```python
df.isnull().sum()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# V biu  phn tn
plt.figure(figsize=(10, 6))
plt.scatter(df['V1'], df['V2'], color='blue', marker='o')

# Thm tiu  v nhn
plt.title('Scatter Plot of V1 vs V2')
plt.xlabel('V1')
plt.ylabel('V2')

# Hin th li
plt.grid()

# Hin th biu 
plt.show()
```

```python
from sklearn.cluster import KMeans
seed = 36
k_cluster = 3
k_mean_model = KMeans(n_clusters= k_cluster, random_state= seed)
x = df[['V1', 'V2']]
k_mean_model.fit(x)
```

```python
labels = k_mean_model.labels_
```

```python
df['Cluster'] = labels
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['V1'], df['V2'], c=df['Cluster'], cmap='viridis', marker='o')
plt.title('K-Means Clustering of V1 and V2')
plt.xlabel('V1')
plt.ylabel('V2')
plt.colorbar(label='Cluster Label')
plt.grid()
plt.show()
```
