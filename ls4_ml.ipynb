```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# Số giờ học
hours_studied = np.array([1, 2, 3, 4, 5], dtype=np.float32)
# Điểm thi
exam_scores = np.array([60, 65, 70, 75, 80], dtype=np.float32)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
model.fit(hours_studied, exam_scores, epochs=500)
```

```python
print(model.predict(np.array([6, 7, 8, 9])))
```

```python
predicted_scores = model.predict(hours_studied)
plt.scatter(hours_studied, exam_scores, color='blue', label='Actual')
plt.plot(hours_studied, predicted_scores, color='red', label='Predicted')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.legend()
plt.show()
```
